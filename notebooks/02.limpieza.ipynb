```python
import pandas as pd
df_raw = pd.read_csv('../data/data_raw.csv')
df_raw.sample(5)
```

### COPIA DEL DATAFRAME ORIGINAL ###

```python
df = df_raw.copy()
```

### CAMBIO NOMBRES DE COLUMNAS ###

```python
df.columns = df.columns.str.replace(' ','_').str.lower()
df.sample(5)
```

### CAMBIAR VISIT_DATE A FECHAS ###

```python
df['visit_date'] = pd.to_datetime(df['visit_date'], format='%Y-%m-%d')
df.info()
```

### crear una columna nueva ###

```python
df['conversion_str'] = df['conversion'].apply(lambda x: 'yes' if x == 1 else 'no')
df.sample(10)    
```

### sacar solo unas columnas ###

```python
df.loc[:,['conversion','conversion_str']]
```

### cambia a MINUSCULAS ###

```python
df['product_category']= df['product_category'].str.lower()
```

### CAMBIAR FASHHION por FASHION ###

```python
df['product_category']= df['product_category'].str.replace('fashhion','fashion')
df['product_category'].value_counts()   
```
