### se ha comprobado que los nulos de payment, shipment y coupon son debidos a que no se ha realizado comora ###

```python
df['payment_method']  = df.loc[df['conversion'] == 0, 'payment_method'].fillna('no_conversion')
df['shipping_method'] = df.loc[df['conversion'] == 0, 'shipping_method'].fillna('no_conversion')
df['coupon_used'] = df.loc[df['conversion'] == 0, 'coupon_used'].fillna('no_conversion')
df.sample(5)
```

```python
df['payment_method'] = df.loc[df['payment_method'].isnull(), 'payment_method'].fillna('no_conversion')
df.isnull().mean()*100
df.sample(5)
```

### convertir tipo de campo a float ###

```python
df_nulos['price'] = df_nulos['price'].astype(str).str.replace(',', '.', regex=False)
df_nulos['price'] = df_nulos['price'].replace(r"<class '.*'>", pd.NA, regex=True)
df_nulos['price'] = pd.to_numeric(df_nulos['price'], errors='coerce')
```
